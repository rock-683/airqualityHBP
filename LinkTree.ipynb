## Feature Selection 
<font color='grey'>*Feature selection for the colon cancer GWR/MGWR model involved using correlation to aid in the initial reduction of variables and Ordinary Least Squares (OLS) regression to evaluatestrength of explanatory variables andto aid in the reduction of variables.*</font> 


6. [Correlation2.ipynb](../SpIn_5_Artifacts/Correlation2.ipynb) This notebook looks at the correlation between varibles in the three key variable groups: Cancer, Nitrogen and County Health Indicators. This scope is the midwest. The correlation was used to evaluate strength of explanatory variables and to aid in the initial reduction of variables. The four variables that emerged with the strongest relationship to colon cancer were Limited access to healthy foods (r=0.34), Infant mortality (0.29), Air pollution (-0.23), and percent Rural (0.21)


7. [Regression_ColonCancer2.ipynb](../SpIn_5_Artifacts/Regression_ColonCancer2.ipynb) 
In this notebook, ordinary least squares regression is used to evaluate strength of explanatory variables and to aid in the reduction of variables. The OLS Regression based off of correlation results in Correlation2 notebook listed above. The four variables that provided the highest explanitory strength for colon cancer rate were the following independent/predictor variables: 
**`NitrogenRate`, `PercentRural`,`LimitedAccessHealthyFoods`,** and **`AirPollution`**.
